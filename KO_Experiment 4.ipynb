# Experiment 4: Data Wrangling and Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

###   ECE BOARD EXAM PROBLEM

#### 1.

```python
#a.
#read the board_excel file.
board=pd.read_excel('board2.xlsx')

#Sort the board to track only data where 'Track' is Instrumentation and 'Hometown' is Luzon.
instru=board[(board['Track'] == 'Instrumentation') & (board['Hometown'] == 'Luzon')]

#Sort the board again to only show data where the 'Electronics' is greater than 70. It will only preview the 'Name','GEAS', and'Electronics'.
instru=instru.loc[(instru['Electronics']>70),['Name','GEAS','Electronics']]

#Save the DataFrame  as an excel file and name it 'Instru'
instru.to_excel('Instru.xlsx',index=False)
```

```python
#b.
#read the board_excel file.
board=pd.read_excel('board2.xlsx')

#get the average of 'Math', 'GEAS', 'Electronics', 'Communication' and set each average in a row.
board['Average']=board[['Math', 'GEAS', 'Electronics', 'Communication']].mean(axis=1)

#Sort the board to track only data where 'Hometown is Mindanao and 'Gender' is Female.
Mindy=board[(board['Hometown'] == 'Mindanao') & (board['Gender'] == 'Female')]

#Sort the board again to only show data where the 'Average' is greater than or euqal to 5. It will only preview the 'Name','GEAS','Electronics', 'Average'.
Mindy=Mindy.loc[(Mindy['Average']>=55),['Name','GEAS','Electronics','Average']]

#Save the DataFrame  as an excel file and name it 'Mindy'
Mindy.to_excel('Mindy.xlsx',index=False)
```

#### 2.

```python
# Set the style for seaborn
sns.set(style="whitegrid")
sns.set_palette("Set2")

# 1. Create a Boxplot to Plot for average grade by track
# Create the size of the boxplot
plt.figure(figsize=(6, 10))

# input the x and y values of the boxplot
sns.boxplot(x='Track', y='Average', data=board)

# adding style to the Boxplot
plt.title('Average Grade by Track', fontsize=20)
plt.xlabel('Track', fontsize=13)
plt.ylabel('Average', fontsize=15)

x=1
#Show the Boxplot
plt.show()

# 2. Create a Boxplot to Plot for average grade by gender
# Create the size of the boxplot
plt.figure(figsize=(6, 10))

# input the x and y values of the boxplot
sns.boxplot(x='Gender', y='Average', data=board)

# adding style to the Boxplot
plt.title('Average Grade by Gender', fontsize=20)
plt.xlabel('Gender', fontsize=13)
plt.ylabel('Average', fontsize=15)

#Show the Boxplot
plt.show()

# 3.Create a Boxplot to Plot for average grade by hometown
# Create the size of the boxplot
plt.figure(figsize=(6, 10))

# input the x and y values of the boxplot
sns.boxplot(x='Hometown', y='Average', data=board)

# adding style to the Boxplot
plt.title('Average Grade by Hometown', fontsize=20)
plt.xlabel('Hometown', fontsize=13)
plt.ylabel('Average', fontsize=15)

#Show the Boxplot
plt.show()
```
